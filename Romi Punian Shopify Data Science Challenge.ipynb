## Question 1: Given some sample data, write a program to answer the following: click here to access the required data set

On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 

- It appears there is a few abnormally large orders that have significantly larger order amounts

- One such example is order ID 16 which is for 2000 sneakers by user ID 607, for and order amount of $704000. It appears user 607 is likely a reseller who buys sneakers in bulk and sells them later at a higher prices as they have multiple large orders. Outliers such as this are probably skewing the AOV if we were to take a mean. 

- There are also orders (such as order ID 161, or 512 for example) with smaller total items ordered that still have large values for their order amounts so these should possibly be investigated further as these are also affecting the average order value.

What metric would you report for this dataset?

- I believe looking for the median of the order amount would be a better metric to look at. First we will have to organize the data by order amount in ascending order.



What is its value?

- After using pandas to sort the data frame by order amount and then finding the median I found the median to be $284.00






```python
# Finding the median

df = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv")

df = df.sort_values(['order_amount'])

dfMed = df['order_amount'].median()

display(df)

display(dfMed)
```

```python
# Finding the mean without large total item orders (less than 20)
# Thought this could be a way of calculating the mean without the outliers of massive orders, 
# however this requires manipulation of the data which may skew the result. Therefore the Median is a better metric.

df = df[df['total_items'] <= 20]

dfMean = df['order_amount'].mean()

#display(df)

display(dfMean)
```

### What product was ordered the most by customers in Germany?

- SQL Queries:

Query to show highest sum of quantity's of all products ordered from Germany in descending order:

SELECT OrderDetails.ProductID, SUM(Quantity) FROM [OrderDetails] WHERE OrderID IN (SELECT OrderID from Orders WHERE CustomerID In (Select CustomerID from Customers WHERE Country = 'Germany')) GROUP BY ProductID ORDER BY SUM(Quantity) DESC;

The above query tells us the product with the highest quantity ordered by German customers is productID 40.

Query to Find ProductID 40 from the products table:

SELECT * FROM [Products] WHERE ProductID = 40;

- Answer: We find that the product with the highest Quantity ordered from customers in Germany was ProductID 40, which is Boston Crab Meat
